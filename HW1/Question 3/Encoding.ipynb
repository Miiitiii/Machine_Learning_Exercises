```python
import numpy as np
from numpy.linalg import inv
from matplotlib import pyplot as plt
import pandas as pd
from numpy import argmax

address = 'D:/HomeWork/Machine Learning/ML-HW01/ML-HW01/dataset3/train.csv'


data = pd.read_csv(address)

data.head(10)
```

```python
for i in range(len(data)):
    if data.at[i , 'gender'] == 'male':
        data.at[i ,'gender'] = 1
    else:
        data.at[i ,'gender'] = 0
```

```python
for i in range(len(data)):
    if data.at[i , 'smoker'] == 'yes':
        data.at[i ,'smoker'] = 1
    else:
        data.at[i ,'smoker'] = 0
```

```python
regions = data['region']

### Universal list of colors
total_regions = ["southwest", "northwest", "northeast", "southeast"]

### map each color to an integer
mapping = {}
for x in range(len(total_regions)):
    mapping[total_regions[x]] = x

one_hot_encode = []

for c in regions:
    arr = list(np.zeros(len(total_regions), dtype = int))
    arr[mapping[c]] = 1
    one_hot_encode.append(arr)
    
for i in range(len(one_hot_encode)):
    data.at[i , 'region'] = one_hot_encode[i]
```
