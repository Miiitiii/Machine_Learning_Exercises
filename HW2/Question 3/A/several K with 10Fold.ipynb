```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from random import randrange
from sklearn.metrics import confusion_matrix

# reading csv files
data =  pd.read_csv('D:/HomeWork/iris.data', sep=",",header = None)

for i in range(len(data)):
    if data.at[i , 4] == 'Iris-setosa':
        data.at[i ,4] = 0
    elif data.at[i , 4] == 'Iris-versicolor':
        data.at[i ,4] = 1
    elif data.at[i , 4] == 'Iris-virginica':
        data.at[i ,4] = 2

print(data)
```

```python
class KNN:
    def __init__(self , X , Y , k):
        self.X_train = X
        self.Y_train = Y
        self.k = k
        
    def euclidean_distance(self , inst1 , inst2):
        distance = np.linalg.norm(inst1 - inst2)
        return distance
    
    
    def get_k_neighbors(self , inst1):
        distances=[]
        
        for inst2 in self.X_train:
            distances.append(self.euclidean_distance(inst1 , inst2))
            
        distances = np.asarray(distances)
        indices = np.argpartition(distances , self.k)
        k_self_indices = indices[:self.k]
        return k_self_indices
    
    
    def predict_class(self , inst1):
        
        knn_indices = self.get_k_neighbors(inst1)
        knn_labels= []
        
        for i in knn_indices:
            knn_labels.append(self.Y_train[i])
            
        occurrences = np.bincount(knn_labels)
        mode = np.argmax(occurrences)
        return mode
    
    def get_accuracy(self , Y_test , prediction):
        correct = Y_test == prediction
        acc = (np.sum(correct) / Y_test.shape[0])*100.0
        return acc
```

```python
#insert data in arrays
y = data.iloc[:, 4].values
x = data.iloc[:, 0:4].values
```

```python
#Cross Validation 10-Fold
k = [1 , 3 , 5 , 7 ,9 , 11 , 13 , 17 , 21 , 23 , 25 , 27 , 29 , 31 , 33 , 35 , 37 , 39 , 41 , 43 , 45 , 47 , 49 , 51]

for K in k :
    accuracy = []
    y_predicts = []
    for i in range(10):
        X = np.split(x, 10)
        Y = np.split(y, 10)
        
        X_Test = X[i]
        Y_Test = Y[i]
        
        X.pop(i)
        Y.pop(i)
        
        classifier = KNN(np.concatenate(X),np.concatenate(Y), K)
        preds = []
        for inst in X_Test :
            preds.append(classifier.predict_class(inst))
            y_predicts.append(classifier.predict_class(inst))
        accuracy.append(classifier.get_accuracy(Y_Test , preds))
    print('For K = ',K , 'Average of Validations ' , np.mean(accuracy))
    print('Confusion Matrix : \n' , confusion_matrix(y.tolist() , y_predicts))
```
