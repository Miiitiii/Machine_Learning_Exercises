```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import tree
from chefboost import Chefboost as chef


# reading csv files
data =  pd.read_csv('D:/HomeWork/Machine Learning/titanic/train.csv')
# reading csv files
dataTest =  pd.read_csv('D:/HomeWork/Machine Learning/titanic/test.csv')

# data.pop('Name')
dfTest = dataTest.copy()
# dfTest.pop('Name')

d = {0: 'NO' , 1:'YES'}

data['Survived'] = data['Survived'].map(d)


data.head()
```

```python
config = {'algorithm' : 'C4.5'}
```

```python
model = chef.fit(data , config, target_label = 'Survived')
```

```python
predict = []
for i in range(len(dataTest)):
    predict.append(chef.predict(model , dfTest.iloc[i]))
```

```python
dataTest['Survived'] = predict
```
