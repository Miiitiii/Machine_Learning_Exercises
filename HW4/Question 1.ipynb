```python
import os
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
import cv2
```

```python
def load_images(directory):
    one_dim = []
    class_names = []
    labels = []
    for filename in os.listdir(directory):
        path = directory + "/" + filename
        t = filename.split('_')
        t = t[0]
        if t not in class_names:
            class_names.append(t)
        labels.append(t)
        im = mpimg.imread(path)
        temp = []
        for row in im:
            for w in row:
                temp.append(w)
        one_dim.append(temp)

    return one_dim, labels
```

```python
trainData , trainlabels = load_images(r"D:\HomeWork\Machine Learning\ML_HW04\ML_HW04\ML_HW04\data\2")
```

```python
#f for class 0 and s for class 1
def kmeanscluster(X , k , mu , maxIter):
    iteration = 0
    means = list()
    
    while iteration < maxIter:
        
        if iteration != 0:
            mu = means
        
        
        newDatas = list()
        
        for item in range(k):
            dist = []
            for i in range(len(X)):
                dist.append(np.linalg.norm(X[i] - mu[item]))
            newDatas.append(np.array(dist))
            
        
        newLabels = np.argmin(np.array(newDatas), axis=0)
        
        means.clear()
        for item in range(k):
            arrangedData = []
            for i in range(len(X)):
                if newLabels[i] == item:
                    arrangedData.append(X[i])
            data = np.array(arrangedData)
            meanData = data.mean(axis = 0)
            means.append(meanData)
            
        iteration = iteration + 1    
    
    
    newImage = list()
    
    for label in newLabels:
        for center in range(len(mu)):
            if label == center:
                newImage.append(mu[center])
                
    return newLabels , mu , newImage
```

```python
def findMu(k , data):
    mu = list()
    for i in range(0,k):
        mu.append(data[np.random.choice(len(data))])
    return np.array(mu)
```

```python
def clustering(k , data):
    mu = findMu(k , data)
    c , centers , newImage = kmeanscluster(data , k , mu , 10)
    percent=[]
    c = c.tolist()
    for i in range(len(centers)):
        j=c.count(i)
        j=j/(len(c))
        percent.append(j)
    plt.pie(percent,colors=np.array(np.array(centers)/255),labels=np.arange(len(centers)))
    plt.show()
    return newImage
```
