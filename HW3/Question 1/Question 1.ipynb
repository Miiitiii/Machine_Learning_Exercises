```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd


# reading csv files
data =  pd.read_csv(r'D:\HomeWork\Machine Learning\ML_HW03\iris.data', sep=",",header = None)

setosa = data[data[4] == 'Iris-setosa']
virginica = data[data[4] == 'Iris-virginica']
versicolor = data[data[4] == 'Iris-versicolor']


print(data)
```

```python
#Feature 1 and 2
F1_setosa = setosa[0]
F2_setosa = setosa[1]

F1_virginica = virginica[0]
F2_virginica = virginica[1]

F1_versicolor = versicolor[0]
F2_versicolor = versicolor[1]


plt.plot(F1_setosa,F2_setosa , linestyle='', marker='o', color="red", label='setosa')
plt.plot(F1_virginica,F2_virginica , linestyle='', marker='o', color="blue", label='virginica')
plt.plot(F1_versicolor,F2_versicolor , linestyle='', marker='o', color="black", label='versicolor')
plt.title('Features : sepal length and sepal width')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.legend()

plt.show()
```

```python
#Feature 3 and 4
F1_setosa = setosa[2]
F2_setosa = setosa[3]

F1_virginica = virginica[2]
F2_virginica = virginica[3]

F1_versicolor = versicolor[2]
F2_versicolor = versicolor[3]


plt.plot(F1_setosa,F2_setosa , linestyle='', marker='o', color="red", label='setosa')
plt.plot(F1_virginica,F2_virginica , linestyle='', marker='o', color="blue", label='virginica')
plt.plot(F1_versicolor,F2_versicolor , linestyle='', marker='o', color="black", label='versicolor')
plt.title('Features : petal length and petal width')
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.legend()

plt.show()
```
