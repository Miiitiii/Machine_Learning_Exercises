```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from sklearn import svm, datasets
from sklearn.metrics import accuracy_score,f1_score
from sklearn.model_selection import train_test_split


# reading csv files
df =  pd.read_csv(r'D:\HomeWork\Machine Learning\ML_HW03\iris.data', sep=",",header = None)
#suffle data
data = df.sample(frac=1)

d = {'Iris-setosa': 0 , 'Iris-virginica':1 , 'Iris-versicolor':2}

data[4] = data[4].map(d)


X_train, X_test, y_train, y_test = train_test_split(data.iloc[: , 2:4], data.iloc[: , 4], test_size=0.30)


#Assumed you have, X (predictor) and Y (target) for training data set and x_test(predictor) of test_dataset
X = X_train.values
y = y_train.values

X_test = X_test.values
y_test = y_test.values

print(data)
```

```python
# Create SVM classification object
def calculateSVM(Type , co , d):
    svc = svm.SVC(kernel=Type, C = co , degree = d).fit(X, y)
    return svc , svc.predict(X_test)
```

```python
#Calculate for Linear Type
```

```python
#call svm calculator with input 0.01
svc , predicted = calculateSVM('linear' , 0.01,1)


# create a mesh to plot in

x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
np.arange(y_min, y_max, h))



Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)



plt.contourf(xx, yy, Z,  alpha=0.4)
plt.scatter(X[:, 0], X[:, 1], c=y)

plt.xlabel('petal length')
plt.ylabel('petal width')

plt.xlim(xx.min(), xx.max())
plt.title('SVC with linear kernel')

plt.show()

print('Accuracy for linear 0.01 is :',accuracy_score(y_test , predicted) *100 ,'%')
print('F1-Score for linear 0.01 is :',f1_score(y_test , predicted, average='macro') *100 ,'%')
```
